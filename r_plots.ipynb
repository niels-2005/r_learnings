# Introduction Plotting with R

```r
library(data.table)

dt <- fread("data/cars.csv")

dt[MPG == 0, MPG := NA]

dt[, `:=` (
  Displacement = NULL, 
  Origin = as.factor(Origin), 
  Weight = Weight / 2.205, 
  l100km = 235.215 / MPG, 
  MPG = NULL
)]

# x = Weight, y = l100km
plot(dt$Weight, dt$l100km, type = "p", xlab = "Gewicht", ylab = "Verbrauch l/100km")
```

```r
library(ggplot2)

dt <- fread("data/cars.csv")

dt[MPG == 0, MPG := NA]

dt[, `:=` (
  Displacement = NULL, 
  Origin = as.factor(Origin), 
  Weight = Weight / 2.205, 
  l100km = 235.215 / MPG, 
  MPG = NULL
)]

p <- qplot(dt$Weight, dt$l100km, xlab = "Gewicht", ylab = "Verbrauch l/100km")
print(p)
```

```r
dt <- fread("data/cars.csv")

dt[MPG == 0, MPG := NA]

dt[, `:=` (
  Displacement = NULL, 
  Origin = as.factor(Origin), 
  Weight = Weight / 2.205, 
  l100km = 235.215 / MPG, 
  MPG = NULL
)]

g <- ggplot(dt, mapping = aes(Weight, l100km)) +
  geom_point(mapping = aes(color = Origin, size = Origin, shape = Origin), 
             alpha = 0.6, na.rm = T) +
  scale_size_manual(values = c(1, 2, 3)) +
  scale_color_manual(values = c("#FF0000", "#FFFF00", "#FF00FF")) + 
  scale_shape_manual(values = c(14, 10, 5))
print(g)
```

### Label graphics

```r
dt <- fread("data/cars.csv")

dt[MPG == 0, MPG := NA]

dt[, `:=` (
  Displacement = NULL, 
  Origin = factor(Origin, levels = c("Europe", "Japan", "US")), 
  Weight = Weight / 2.205, 
  l100km = 235.215 / MPG, 
  MPG = NULL
)]

g <- ggplot(dt, mapping = aes(Weight, l100km)) +
  geom_point(mapping = aes(color = Origin, shape = Origin), 
             alpha = 0.6, na.rm = T) +
  scale_size_manual(values = c(1, 2, 3)) +
  scale_color_discrete(labels = c("Europa", "Japan", "USA")) +
  scale_shape_discrete(labels = c("Europa", "Japan", "USA")) + 
  labs(title = "Verbrauch von Autos", color = "Herkunft", shape = "Herkunft") + 
  xlab("Gewicht") + 
  ylab("Verbrauch pro 100km")

print(g)
```

### more than one plot

```r
dt <- fread("data/cars.csv")

dt[MPG == 0, MPG := NA]

dt[, `:=` (
  Displacement = NULL, 
  Origin = as.factor(Origin), 
  Weight = Weight / 2.205, 
  l100km = 235.215 / MPG, 
  MPG = NULL
)]

# Grafik erstellen

g <- ggplot(dt, mapping = aes(Weight, l100km)) +
  geom_point(mapping = aes(color = Origin), alpha = 0.6, na.rm = T) + 
  facet_wrap(vars(Origin), ncol = 1)
print(g)
```

### annotate

```r
library(ggrepel)

dt <- fread("data/cars.csv")

dt[MPG == 0, MPG := NA]

dt[, `:=` (
  Displacement = NULL, 
  Origin = as.factor(Origin), 
  Weight = Weight / 2.205, 
  l100km = 235.215 / MPG, 
  MPG = NULL
)]

dt.sample <- dt[sample(.N, 25)]

# Grafik erstellen
g <- ggplot(dt.sample, mapping = aes(Weight, l100km)) +
  geom_point(mapping = aes(color = Origin), alpha = 0.6, na.rm = T) +
  geom_label_repel(mapping = aes(label = Car, color = Origin), na.rm = T, size = 3)
  
print(g)
```

### Barplots

```r
dt <- fread("data/cars.csv")

dt[MPG == 0, MPG := NA]

dt[, `:=` (
  Displacement = NULL, 
  Origin = as.factor(Origin), 
  Weight = Weight / 2.205, 
  l100km = 235.215 / MPG, 
  MPG = NULL
)]

# geom_bar = Barplot
g <- ggplot(dt) +
  geom_bar(aes(y = Origin, fill = as.factor(Cylinders)), colour = "black") +
  labs(fill = "Zylinder")
print(g)
```

### Histograms

```r
dt <- fread("data/cars.csv")

dt[MPG == 0, MPG := NA]

dt[, `:=` (
  Displacement = NULL, 
  Origin = as.factor(Origin), 
  Weight = Weight / 2.205, 
  l100km = 235.215 / MPG, 
  MPG = NULL
)]

# geom_histogram = Histogram
g <- ggplot(dt) +
  geom_histogram(aes(x = l100km, fill = Origin), binwidth = 10, center = 5, na.rm = T)
print(g)
```
