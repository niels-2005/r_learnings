### group by

```r
dt <- fread("data/cars.csv")

dt[MPG == 0, MPG := NA]

dt[, `:=` (
  Displacement = NULL, 
  Origin = as.factor(Origin), 
  Weight = Weight / 2.205, 
  l100km = 235.215 / MPG, 
  MPG = NULL
)]

# keyby is like groupby
grouped <- dt[!is.na(l100km), .(mean = mean(l100km), min = min(l100km), mean_kg = mean(Weight)), keyby = Origin]

print(grouped)
```

### sorting

```r
dt <- fread("data/cars.csv")

dt[MPG == 0, MPG := NA]

dt[, `:=` (
  Displacement = NULL, 
  Origin = as.factor(Origin), 
  Weight = Weight / 2.205, 
  l100km = 235.215 / MPG, 
  MPG = NULL
)]

# sorting order
setorder(dt, -Origin, -l100km)

print(dt)
```

### unique

```r
dt <- fread("data/cars.csv")

dt[MPG == 0, MPG := NA]

dt[, `:=` (
  Displacement = NULL, 
  Origin = as.factor(Origin), 
  Weight = Weight / 2.205, 
  l100km = 235.215 / MPG, 
  MPG = NULL
)]

# get unique numbers
print(uniqueN(dt, by = c("Origin", "Cylinders")))
```

### set names

```r
dt <- fread("data/cars.csv")

dt[MPG == 0, MPG := NA]

dt[, `:=` (
  Displacement = NULL, 
  Origin = as.factor(Origin), 
  Weight = Weight / 2.205, 
  l100km = 235.215 / MPG, 
  MPG = NULL
)]

# change names from columns
setnames(dt, c("Car", "Horsepower"), c("Name", "HP"))

print(dt)
```
